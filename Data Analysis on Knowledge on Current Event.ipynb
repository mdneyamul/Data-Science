#**5. Calculate and Explain the statistics (mean/median/mode, range/variance/standard deviationetc.) of each feature.**<br>
**Answer:<br>**
**Mean:** The mean (average) of a data set is found by adding all numbers in the data set and then dividing by the number of values in the set.

```
#@title Calculate mean/median/mode, range/variance/standard deviation 

print("Mean\n\n",df.mean())
print("--------------------------------------------------------\n\n")
#Median
print("Median\n")
print(df.median())
print("--------------------------------------------------------\n\n")
#Mode
print("Mode\n")
print("Age : ",df['Age'].mode().iat[0])
print("Informed about Breonna Taylor Case : ",df['Informed about Breonna Taylor Case'].mode().iat[0])
print("Informed about Explosion in Lebanon : ",df['Informed about Explosion in Lebanon'].mode().iat[0])
print("Informed about Supreme Court  Nomination in US  : ",df['Informed about Supreme Court  Nomination in US'].mode().iat[0])
print("--------------------------------------------------------\n\n")

#Range
print("Range\n")

print("Age : ",df['Age'].max()-df['Age'].min())
print("Informed about Breonna Taylor Case : ",df['Informed about Breonna Taylor Case'].max()-df['Informed about Breonna Taylor Case'].min())
print("Informed about Explosion in Lebanon : ",df['Informed about Explosion in Lebanon'].max()-df['Informed about Explosion in Lebanon'].min())
print("Informed about Supreme Court  Nomination in US  : ",df['Informed about Supreme Court  Nomination in US'].max()-df['Informed about Supreme Court  Nomination in US'].min())
print("--------------------------------------------------------\n\n")
#Varience
print("Varience\n\n",df.var())
print("--------------------------------------------------------\n\n")
#Standard deviation
print("Standard Deviation\n\n",df.std())
```
